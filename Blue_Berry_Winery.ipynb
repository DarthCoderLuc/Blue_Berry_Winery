```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
df_red_wine = pd.read_csv("/Users/Lucasmeissner/Desktop/Blue_Berry_Winery/wine+quality/winequality-red.csv", sep= ';')
df_red_wine
```

```python
df_white_wine = pd.read_csv("/Users/Lucasmeissner/Desktop/Blue_Berry_Winery/wine+quality/winequality-white.csv", sep= ';')
df_white_wine
```

```python
df_red_wine.sort_values(by=['quality'],ascending=False).head(100)
```

```python
df_red_wine['quality_label'] = df_red_wine['quality'].apply(lambda value: 'low'
if value <= 5 else 'medium'
if value <= 7 else 'high')
```

```python
df_red_wine
```

```python
df_white_wine['quality_label']= df_white_wine['quality'].apply(lambda value: 'low'
if value <= 5 else 'medium'
if value <= 7 else 'high')
```

```python
df_all_wines= pd.concat([df_red_wine[['residual sugar','total sulfur dioxide','sulphates','alcohol','volatile acidity','quality','quality_label']].describe(), df_white_wine[['residual sugar','total sulfur dioxide','sulphates','alcohol','volatile acidity','quality','quality_label']].describe()], axis=1, keys=["Red Wines Stats", "White Wines Stats"])
df_all_wines


```

```python
pd.concat([df_red_wine[['residual sugar','quality',]].describe(), df_white_wine[['residual sugar','quality',]].describe()], axis=1, keys=["Red Wines Stats", "White Wines Stats"])

#1 key takeaway is that white wine has a much wider range in sugar residual than red wine due to standard deviation and min/max values
#2 in general all white wines have much higher sugar residual in average than red wines
```

```python
pd.concat([df_red_wine[['total sulfur dioxide','quality']].describe(), df_white_wine[['total sulfur dioxide','quality']].describe()], axis=1, keys=["Red Wines Stats", "White Wines Stats"])

#legal limits for sulfur dioxides for red wine
#EU = 160mg/L

#legal limits for sulfur dioxides for white wine and rose 
#EU = 210mg/L

#legal limits for sulfur dioxides for sweet wines
#EU = 400mg/L
#US = 350mg/L
#Australia = 250mg/L

#key takeaways
#1 there are legal limits which need to be considered for different regions
#2 there are white and red wine(s) in our dataset which exceed the legal limit of 210mg/L or 160mg/L
#3 need to categorize wine into sweet/rose etc. in order to match them correctly
```

```python
pd.concat([df_red_wine[['sulphates','quality']].describe(), df_white_wine[['sulphates','quality']].describe()], axis=1, keys=["Red Wines Stats", "White Wines Stats"])

### sulfates necessary to make wine but high sulfate is considered to be linked with headache/hangover
### hence the less solphates the higher the quality of wine
###red wine has slightly higher mean in sulphates than white wine and also has the wine with the most sulphates (max) + slightly higher std 
```

```python
pd.concat([df_red_wine[['alcohol','quality']].describe(), df_white_wine[['alcohol','quality']].describe()], axis=1, keys=["Red Wines Stats", "White Wines Stats"])

### on average (mean) red and white are almost identical in regards to alcohol
### white wine has a slightly higher std
```

```python
pd.concat([df_red_wine[['volatile acidity','quality']].describe(), df_white_wine[['volatile acidity','quality']].describe()], axis=1, keys=["Red Wines Stats", "White Wines Stats"])

### process of turning wine into vinegar
### legal limits in US:
### max. 1.2 G/L for red
### max. 1.1 G/L for white

###red wine has on average almost twice as much v.a.
```

```python
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Red Wine', 'White Wine'
sizes = [1599, 4898]
explode = (0, 0.1)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
